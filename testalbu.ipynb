```python
import albumentations as albu
import cv2
import torch
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('castom-masks/train/img/AAE_0004.png')
mask = cv2.imread('castom-masks/train/masks_machine/AAE_0004.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
plt.imshow(mask*255)
```

```python
transform = [
    albu.augmentations.RandomCrop(1026,1026,p=0.5),
    albu.augmentations.transforms.GaussNoise(always_apply=True),
    albu.augmentations.transforms.RandomBrightness(always_apply=True),
    albu.augmentations.transforms.RandomToneCurve(always_apply=True),
    albu.augmentations.transforms.RandomGamma(always_apply=True)
    ]
aug = albu.Compose(transforms=transform)
```

```python
sample = aug(image = image,mask = mask)
```

```python
plt.imshow(sample['image'])
plt.figure()
plt.imshow(sample['mask']*255)
```
